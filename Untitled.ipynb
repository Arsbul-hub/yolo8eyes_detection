```python
from ultralytics import YOLO

```

```python

# Load a pretrained YOLO26n model
model = YOLO("./d.pt")

# Train the model on the COCO8 dataset for 100 epochs
```

```python
train_results = model.train(
    data="./trainv2/data.yaml",  # Path to dataset configuration file
    epochs=1,  # Number of training epochs
    imgsz=1024,  # Image size for training
    device="cpu",  # Device to run on (e.g., 'cpu', 0, [0,1,2,3])
)

```

```python

# Evaluate the model's performance on the validation set
metrics = model.val()
```

```python
# Perform object detection on an image
results = model("./trainv2/train/images/1b8f73dd-00dacaae3139e2af.jpg")  # Predict on an image
results[0].show()  # Display results

```

```python
import cv2
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Ошибка: не удалось открыть видеопоток")
    exit()

# 3. Настройки для детекции (опционально)
# Можно передавать параметры прямо в model() или задать глобально:
# model.conf = 0.25   # порог уверенности
# model.iou = 0.45    # порог NMS

while True:
    # 4. Чтение кадра
    ret, frame = cap.read()
    if not ret:
        print("Поток завершён")
        break


    results = model(frame)  # можно добавить параметры: model(frame, conf=0.25, iou=0.45)
   

    #    Метод plot() возвращает изображение в формате numpy (BGR)
    annotated_frame = results[0].plot()

    cv2.imshow("YOLOv8 Detection", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 9. Освобождаем ресурсы
cap.release()
cv2.destroyAllWindows()
```

```python
# Export the model to ONNX format for deployment
path = model.export(format="onnx")  # Returns the path to the exported model
```
